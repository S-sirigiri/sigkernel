```python
import base

import numpy as np
import torch
import time
import pandas as pd
from tqdm.notebook import trange

from tools import brownian, truncated_sigKernel, naive_sig_kernel, generate
from sigKer_fast import sig_kernel

import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
params = {'legend.fontsize': 14,
          'figure.figsize': (16, 16),
          'axes.labelsize': 14,
          'axes.titlesize': 14,
          'xtick.labelsize': 14,
          'ytick.labelsize': 14}
pylab.rcParams.update(params)
```

# Error Analysis

```python
# generate data
length_x, length_y, hurst_x, hurst_y, depth, dim = 100, 100, .8, .8, 10, 3

x,y = generate(length_x, length_y, hurst_x, hurst_y, BM=True, dim_BM=dim, scale=1.)
```

```python
# naive computation
k_true = naive_sig_kernel(x,y,depth)
```

```python
# PDE computations
kernels_Goursat = []
max_n = 3
for n in range(max_n):
    k_Goursat = sig_kernel(x, y, n=n, solver=1, full=True)
    kernels_Goursat.append(k_Goursat[::2**n, ::2**n])
```

```python
r = 16
colorbar_range = np.linspace(min(k_true.min(), kernels_Goursat[-1].min()), max(k_true.max(), kernels_Goursat[-1].max()), 10).tolist()

fig, ax = plt.subplots(1, max_n, figsize=(16,5), sharey=True)

axes = [ax[k].contourf(np.abs(k_true-kernels_Goursat[k]), cmap='seismic') for k in range(max_n)]

# ax0 = ax[0].contourf(k_true, colorbar_range, cmap='RdYlBu')

for k in range(max_n):
    fig.colorbar(axes[k], ax=ax[k])
    ax[k].set_title(f'PDE solver error on grid $P_{k}$', fontsize=r)
    ax[k].set_xlabel('time step s', fontsize=r)
    
# ax[0].set_title('SigKernel $k_{x,y}(s,t)$', fontsize=r)

ax[0].set_ylabel('time step t', fontsize=r)

plt.savefig('../pictures/countour.png')
plt.tight_layout()
plt.show()
```

# Time complexity

```python
depth = 5
```

# Dependence on number of observations

```python
t_naive = []
t_trunc = []
t_pde0 = []
t_pde1 = []
t_pde2 = []

lengths = trange(10, 50, 5)
lengths_ = range(10, 50, 5)
dim = 3
n_times = 5

for M in lengths:
    
    x, y = generate(M,M,BM=True,dim_BM=dim)
    
    t = time.time()
    _ = [naive_sig_kernel(x,y,depth) for k in range(n_times)]
    t_naive.append((time.time()-t)/float(n_times))
    
#     t = time.time()
#     _ = [truncated_sigKernel(np.array([x,y]), depth) for k in range(n_times)]
#     t_trunc.append((time.time()-t)/float(n_times))
    
    t = time.time()
    _ = [sig_kernel(x,y,n=0) for k in range(n_times)]
    t_pde0.append((time.time()-t)/float(n_times))
    
    t = time.time()
    _ = [sig_kernel(x,y,n=1) for k in range(n_times)]
    t_pde1.append((time.time()-t)/float(n_times))
    
    t = time.time()
    _ = [sig_kernel(x,y,n=2) for k in range(n_times)]
    t_pde2.append((time.time()-t)/float(n_times))
```

# Dependence on dimension

```python
t_trunc_dim = []
t_pde_dim0 = []
t_pde_dim1 = []
t_pde_dim2 = []

M = 50
dims = trange(1, 50, 5)
dims_ = range(1, 50, 5)

for dim in dims:
    
    # generate d-dim BM
    x,y = generate(M, M, BM=True, dim_BM=dim)
    
    t = time.time()
    _ = [truncated_sigKernel(np.array([x,y]), depth) for k in range(n_times)]
    t_trunc_dim.append((time.time()-t)/float(n_times))
    
    t = time.time()
    _ = [sig_kernel(x,y,n=0) for k in range(n_times)]
    t_pde_dim0.append((time.time()-t)/float(n_times))
    
    t = time.time()
    _ = [sig_kernel(x,y,n=1) for k in range(n_times)]
    t_pde_dim1.append((time.time()-t)/float(n_times))
    
    t = time.time()
    _ = [sig_kernel(x,y,n=2) for k in range(n_times)]
    t_pde_dim2.append((time.time()-t)/float(n_times))
```

# Visualization of results

```python
fig, ax = plt.subplots(1,2,figsize=(14,5))

ax=plt.subplot(1, 2, 1)
ax.plot(lengths_, t_naive, label=f'truncated (n={depth}) sig-kernel algo [KO19]')
ax.plot(lengths_, t_pde0, label='untruncated sig-kernel PDE on $P_0$')
ax.plot(lengths_, t_pde1, label='untruncated sig-kernel PDE on $P_1$')
ax.plot(lengths_, t_pde2, label='untruncated sig-kernel PDE on $P_2$')

ax.set_title('Dependence on length')
ax.set_xlabel('number of time steps in input paths')
ax.set_ylabel('Elapsed time (s)')
ax.legend()
# ax.set_ylim(0,2.5)
ax.legend(loc="upper left")
ax.grid(True)

ax=plt.subplot(1, 2, 2)
ax.plot(dims_, t_trunc_dim, label=f"truncated (n={depth}) sig-kernel algo [KO19]")
ax.plot(dims_, t_pde_dim0, label='untruncated sig-kernel PDE on $P_0$')
ax.plot(dims_, t_pde_dim1, label='untruncated sig-kernel PDE on $P_1$')
ax.plot(dims_, t_pde_dim2, label='untruncated sig-kernel PDE on $P_2$')
ax.set_title('Dependence on dimension')
ax.set_xlabel('dimension of input paths')
ax.set_ylabel('Elapsed time (s)')
ax.set_ylim(0,4)
ax.legend(loc="upper left")
ax.grid(True)


plt.subplots_adjust(wspace=0.2, hspace=0.4)
plt.tight_layout()
plt.savefig('../pictures/perfomance.png')
plt.show()
```
