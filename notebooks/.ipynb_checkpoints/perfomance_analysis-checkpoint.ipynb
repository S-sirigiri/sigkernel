```python
def gram_matrix(X,Y):
    G = np.zeros((len(X), len(Y)))
    for i in range(len(X)):
        for j in range(len(Y)):
            G[i,j] = sig_kernel(X[i],Y[j])
    return G
```

```python
def ones(m):
    return np.array(m*[1./float(m)])
```

```python
np.eye(5)
```

```python
def betas(X,X_,epsilon=0.1):
    n = len(X)
    m = len(X_)
    A = gram_matrix(X,X)
    B = gram_matrix(X,X_)
    inv = np.linalg.inv(A+n*epsilon*np.eye(n))
    np.dot(np.matmul(inv,B),ones(m))
```

```python
def generate(M, N, h_x=0.8, h_y=0.8, scale=1., signature=False, BM=False, dim_BM=2):
    
    if BM:
        X = brownian(M-1, dim_BM, time=1.)
        Y = brownian(N-1, dim_BM, time=1.)

    else:
        fbm_generator_X = FBM(M-1, h_x)
        fbm_generator_Y = FBM(N-1, h_y)

        x = scale*fbm_generator_X.fbm()
        y = scale*fbm_generator_Y.fbm()

        X = AddTime().fit_transform([x])[0]
        Y = AddTime().fit_transform([y])[0]
    
    if signature:
        X = iisignature.sig(X,5,2)
        Y = iisignature.sig(Y,5,2)

        X0 = np.zeros_like(X[0,:].reshape(1,-1))
        X0[0,0] = 1.
        X = np.concatenate([X0, X])
        Y = np.concatenate([X0, Y])
        
    return 0.5*X, 0.5*Y
```

# Dependence on number of observations

```python
t_naive = []
t_trunc = []
t_pde0 = []
t_pde1 = []
t_pde2 = []
t_pde3 = []

lengths = range(10, 200, 5)
depth = 2

for M in tqdm.tqdm(lengths):
    
    X, Y = generate(M,M)
    
    t = time.time()
    K_naive = np.ones((M,M))
    for i in range(1,M):
        for j in range(1, M):
            K_naive[i,j] = naive_SigKernel_iisig(X[:i+1], Y[:j+1], depth)
    t_naive.append(time.time()-t)
    
    t = time.time()
    K_trunc = truncated_sigKernel(np.array([X,Y]), depth)
    t_trunc.append(time.time()-t)
    
    t = time.time()
    K_PDE0 = sig_kernel(X,Y,n=0)
    t_pde0.append(time.time()-t)
    
    t = time.time()
    K_PDE1 = sig_kernel(X,Y,n=1)
    t_pde1.append(time.time()-t)
    
    t = time.time()
    K_PDE2 = sig_kernel(X,Y,n=2)
    t_pde2.append(time.time()-t)
    
    t = time.time()
    K_PDE3 = sig_kernel(X,Y,n=3)
    t_pde3.append(time.time()-t)
```

# Dependence on dimension

```python
t_trunc_dim = []
t_pde_dim0 = []
t_pde_dim1 = []
t_pde_dim2 = []
t_pde_dim3 = []

M = 30
dims = range(1, 500, 5)
depth = 2

for dim in tqdm.tqdm(dims):
    
    # generate d-dim BM
    X,Y = generate(M, M, BM=True, dim_BM=dim)
    
    t = time.time()
    K_trunc = truncated_sigKernel(np.array([X,Y]), depth)
    t_trunc_dim.append(time.time()-t)
    
    t = time.time()
    K_PDE0 = sig_kernel(X,Y,n=0)
    t_pde_dim0.append(time.time()-t)
    
    t = time.time()
    K_PDE1 = sig_kernel(X,Y,n=1)
    t_pde_dim1.append(time.time()-t)
    
    t = time.time()
    K_PDE2 = sig_kernel(X,Y,n=2)
    t_pde_dim2.append(time.time()-t)
    
    t = time.time()
    K_PDE3 = sig_kernel(X,Y,n=3)
    t_pde_dim3.append(time.time()-t)
```
