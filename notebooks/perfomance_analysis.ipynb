# Error Analysis

```python
# generate data
M = 500
depth = 10
d = 3
x,y = generate(M,M,BM=True,dim_BM=d)
```

```python
# naive computation
k_true = naive_sig_kernel(x,y,depth)
```

```python
# PDE computations
kernels_Goursat = []
nn = 2
for n in range(nn):
    k_Goursat = sig_kernel(x, y, n=n, full=True)
    kernels_Goursat.append(k_Goursat[::2**n, ::2**n])
```

```python
r = 16
colorbar_range = np.linspace(min(k_true.min(), kernels_Goursat[-1].min()), max(k_true.max(), kernels_Goursat[-1].max()), 10).tolist()

fig, ax = plt.subplots(1, 3, figsize=(16,5), sharey=True)

ax0 = ax[0].contourf(k_true, colorbar_range)
ax1 = ax[1].contourf(np.abs(k_true-kernels_Goursat[0]), cmap='cool')
ax2 = ax[2].contourf(np.abs(k_true-kernels_Goursat[1]), cmap='cool')

fig.colorbar(ax0, ax=ax[0])
fig.colorbar(ax1, ax=ax[1])
fig.colorbar(ax2, ax=ax[2])

ax[0].set_title('Untruncated sig-kernel $k_{x,y}(s,t)$', fontsize=r)
ax[1].set_title('PDE abs error on grid $P_0$', fontsize=r)
ax[2].set_title('PDE abs error on grid $P_1$', fontsize=r)

ax[0].set_xlabel('time step s', fontsize=r)
ax[1].set_xlabel('time step s', fontsize=r)
ax[2].set_xlabel('time step s', fontsize=r)
ax[0].set_ylabel('time step t', fontsize=r)

plt.savefig('../pictures/countour.png')
plt.tight_layout()
plt.show()
```

# Time complexity

```python
depth = 8
```

# Dependence on number of observations

```python
t_naive = []
t_trunc = []
t_pde0 = []
t_pde1 = []
t_pde2 = []

lengths = range(10, 200, 5)
dim = 3

for M in tqdm.tqdm(lengths):
    
    x, y = generate(M,M,BM=True,dim_BM=dim)
    
    t = time.time()
    _ = naive_sig_kernel(x,y,depth)
    t_naive.append(time.time()-t)
    
#     t = time.time()
#     _ = truncated_sigKernel(np.array([x,y]), depth)
#     t_trunc.append(time.time()-t)
    
    t = time.time()
    _ = sig_kernel(x,y,n=0)
    t_pde0.append(time.time()-t)
    
    t = time.time()
    _ = sig_kernel(x,y,n=1)
    t_pde1.append(time.time()-t)
    
    t = time.time()
    _ = sig_kernel(x,y,n=2)
    t_pde2.append(time.time()-t)
```

# Dependence on dimension

```python
t_trunc_dim = []
t_pde_dim0 = []
t_pde_dim1 = []
t_pde_dim2 = []

M = 50
dims = range(1, 500, 5)

for dim in tqdm.tqdm(dims):
    
    # generate d-dim BM
    x,y = generate(M, M, BM=True, dim_BM=dim)
    
    t = time.time()
    _ = truncated_sigKernel(np.array([x,y]), depth)
    t_trunc_dim.append(time.time()-t)
    
    t = time.time()
    _ = sig_kernel(x,y,n=0)
    t_pde_dim0.append(time.time()-t)
    
    t = time.time()
    _ = sig_kernel(x,y,n=1)
    t_pde_dim1.append(time.time()-t)
    
    t = time.time()
    _ = sig_kernel(x,y,n=2)
    t_pde_dim2.append(time.time()-t)
```

# Visualization of results

```python
fig, ax = plt.subplots(1,2,figsize=(14,5))

ax=plt.subplot(1, 2, 1)
ax.plot(lengths, t_naive, label=f'truncated (n={depth}) sig-kernel algo [KO19]')
ax.plot(lengths, t_pde0, label='untruncated sig-kernel PDE on $P_0$')
ax.plot(lengths, t_pde1, label='untruncated sig-kernel PDE on $P_1$')
ax.plot(lengths, t_pde2, label='untruncated sig-kernel PDE on $P_2$')

ax.set_title('Dependence on length')
ax.set_xlabel('number of time steps in input paths')
ax.set_ylabel('Elapsed time (s)')
ax.legend()
# ax.set_ylim(0,2.5)
ax.legend(loc="upper left")
ax.grid(True)

ax=plt.subplot(1, 2, 2)
ax.plot(dims, t_trunc_dim, label=f"truncated (n={depth}) sig-kernel algo [KO19]")
ax.plot(dims, t_pde_dim0, label='untruncated sig-kernel PDE on $P_0$')
ax.plot(dims, t_pde_dim1, label='untruncated sig-kernel PDE on $P_1$')
ax.plot(dims, t_pde_dim2, label='untruncated sig-kernel PDE on $P_2$')
ax.set_title('Dependence on dimension')
ax.set_xlabel('dimension of input paths')
ax.set_ylabel('Elapsed time (s)')
ax.set_ylim(0,4)
ax.legend(loc="upper left")
ax.grid(True)


plt.subplots_adjust(wspace=0.2, hspace=0.4)
plt.tight_layout()
plt.savefig('../pictures/perfomance.png')
plt.show()
```
