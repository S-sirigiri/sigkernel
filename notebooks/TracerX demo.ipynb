# The Signature features of a multivariate time-series

Let $X: [0,T] \to \mathbb{R}^d$ be a path (or time-series, signal, stream etc.). 

For any sub-interval $[a,b] \in [0,T]$ the Signature $S(X)_{[a,b]}$ of the path $x$ over $[a,b]$ is defined as the following collection of coefficients:

\begin{align}
S(X)_{[a,b]} &= \Big(1, \underset{t_1 \in [a,b]}{\int} dX_{t_1}, \ldots, \underset{\substack{t_1<...<t_k \\ t_1,...,t_k \in [a,b]}}{\int ... \int}dX_{t_1}\otimes...\otimes dX_{t_k}, \ldots \Big)
\end{align}

where 

\begin{align}
\underset{t_1 \in [a,b]}{\int} dX_{t_1} = (S_i)_{i=1}^d
\end{align}

is a vector in $\mathbb{R}^d$

\begin{align}
\underset{\substack{t_1<t_2 \\ t_1,t_2 \in [a,b]}}{\int\int}dX_{t_1}\otimes dX_{t_2} = (S_{i,j})_{i,j=1}^d
\end{align}

is a matrix in $\mathbb{R}^{d \times d}$ (a $d \times d$ matrix)

\begin{align}
\underset{\substack{t_1<t_2<t_3 \\ t_1,t_2,t_3 \in [a,b]}}{\int ... \int}dX_{t_1}\otimes dX_{t_2}\otimes dX_{t_3} = (S_{i,j, k})_{i,j,k=1}^d
\end{align}

is a $d \times d \times d $ tensor in $\mathbb{R}^{d \times d \times d}$ 

etc.

### In practice

The signature takes an input 

- a np.array X of shape (length, dim) 
- an integer corresponding to the desired truncation level

and returns a flatten collection of coefficients. Note that the higher the truncation level, the more information is captured. However, the number of features in the signature augments exponentially in the truncation.

### Example

```python
length, dim = 1000, 4
X = brownian(length, dim)
```

```python
plt.figure(figsize=(10,6))
for d in range(dim):
    plt.plot(X[:,d], label=f'stream dimension {d}')
plt.legend()
plt.show()
```

```python
truncation = 2
iisignature.sig(X, truncation)
```

# The Expected Signature features of a sequential tree

### Collections of data-streams

Suppose that instead of dealing with one single data-stream $X$ we deal with a collection of streams $C = \{X_1, \ldots, X_n\}$. Then the Expected Signature plays the same role as the Signature and it is simply defined as 

\begin{align}
    ES(C) = \frac{1}{n} \sum_{i=1}^n S(X_i)
\end{align}

### Sequential trees

A data-tree $t=(v,f)$ is defined recursively as a tuple given by a value $v$ (of some data type, possibly empty) and a forest $f=(t_1, \ldots, t_k)$ (a possibly empty list of trees).

A branching process $P$ consists of a tree $t=(v,f)$ where all values are of the form $v=(w,\gamma)$ where $w \in \mathbb{R}$ is a scalar weight and $\gamma : [0,T] \to \mathbb{R}^d$ is a data-stream of information happening in the given process.

The path-distribution of a branching process $P$ is the collection of all data-streams obtained by concatenating all the data-streams $\gamma$'s along every single subsequent branch of $P$.

### Example
