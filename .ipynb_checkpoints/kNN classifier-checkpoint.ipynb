```python
import matplotlib.pyplot as plt
import seaborn as sns
from knn_classifier import knn_classifier
from tslearn.datasets import UCR_UEA_datasets
```

```python
data = UCR_UEA_datasets().list_multivariate_datasets()
```

```python
x_train, y_train, x_test, y_test = UCR_UEA_datasets().load_dataset(data[2])
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
label, proba, acc_score, conf_mat = knn_classifier(x_train, x_test, y_train, y_test, sub_rate=0.5, metric='soft_dtw',
                                                   length=x_train.shape[1], replace=False, n_neighbours=1)
```

```python
acc_score
```

```python
sns.heatmap(conf_mat, cmap="Greens", annot=True)
plt.show()
```
